```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Marketing Team Data Excelerate.csv")
data
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
data = data.dropna()
data
```

```python
data.shape
```

```python
numeric_columns = data.select_dtypes(include='number')
numeric_columns
```

```python
data.describe()
```

```python
data = data.rename(columns={'A': 'Column 1', 'B': 'Column 2'})
```

```python
# Bar plot for reach vs clicks
plt.figure(figsize=(12, 6))
sns.barplot(x='Reach', y='Clicks', data=data)
plt.title('Clicks by Reach')
plt.xlabel('Reach')
plt.ylabel('Clicks')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for reach vs campain name and age group
plt.figure(figsize=(12, 6))
plt.bar(data['Campaign Name'] + ' - ' + data['Age'], data['Reach'], color='skyblue', label='Reach')
plt.ylabel('Reach')
plt.xlabel('Campaign Name and Age Group')
plt.title('Reach by Campaign and Age Group')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for Click-Through Rate (CTR) vs Cost Per Click (CPC)
plt.figure(figsize=(10, 6))
plt.scatter(data['Click-Through Rate (CTR)'], data['Cost Per Click (CPC)'], c='blue', marker='o')
plt.xlabel('Click-Through Rate (CTR)')
plt.ylabel('Cost Per Click (CPC)')
plt.title('Click-Through Rate vs. Cost Per Click')
plt.grid(True)
plt.show()
```

```python
# Pie chart for Distribution of Clicks by Age Group
clicks_by_age = data.groupby('Age')['Clicks'].sum()
plt.figure(figsize=(8, 8))
plt.pie(clicks_by_age, labels=clicks_by_age.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Clicks by Age Group')
plt.show()
```

```python
numerical_columns = ['Reach', 'Impressions', 'Frequency', 'Clicks', 'Unique Clicks']

corr_matrix = data[numerical_columns].corr()

# Generate the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Between Campaign Metrics')
plt.show()
```
